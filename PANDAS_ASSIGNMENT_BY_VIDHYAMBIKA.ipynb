**Dataset: Diamonds**

Its Attributes are:

* price:	price in US dollars (\$326--\$18,823)
* carat:	weight of the diamond (0.2--5.01)
* cut:	quality of the cut (Fair, Good, Very Good, Premium, Ideal)
* color:	diamond colour, from J (worst) to D (best)
* clarity:	a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
* x:	length in mm (0--10.74)
* y:	width in mm (0--58.9)
* z:	depth in mm (0--31.8)
* depth:	total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)
* table:	width of top of diamond relative to widest point (43--95)

```python
# multi-list
list = [ ['I'], ['am'], ['learning'], ['and'],
         ['experimenting'], ['with'], ['Pandas'], ['for'], ['fun'] ]

# create Pandas Series
series_one = pd.Series((i[0] for i in list))


print(series_one)
```

Here,

* '(i[0] for i in list)' is a generator expression which takes only the first element in each python sub-list.
* Jupyter automatically formats the output to display the entire Series, and you will see the complete string for each element, despite the fact that the generator expression technically only extracts the first element.
* This behavior is specific to Jupyter's interactive environment and its handling of Pandas objects for better visualization in notebooks.
* If you were to print or inspect the series in a non-interactive environment, you might see only the first element.
* You can also use 'series_one = pd.Series([' '.join(i) for i in my_list])'.
* This joins the elements of each sub-list into a single string using a space as a separator. The resulting list is then used to create the Pandas Series.

### **63. Write a Pandas program to get sample 75% of the diamonds Data Frame's rows without replacement and store the remaining 25% of the rows in another DataFrame**

```python
print("Original Dataframe:")
print(diamonds.shape)
print("\nSample 75% of diamonds DataFrame's rows without replacement:")
result = diamonds.sample(frac=0.75, random_state=99)
print(result)
print("\n Remaining 25% of the rows:")
print(diamonds.loc[~diamonds.index.isin(result.index), :])
```

One can also use this code to get the remaining 25% of the rows:

remaining_rows = diamonds.drop(sampled_rows.index)

Let's break down the expression diamonds.loc[~diamonds.index.isin(result.index), :] step by step:

* diamonds.index.isin(result.index):

  * diamonds.index refers to the index of the original DataFrame (diamonds).
  * result.index refers to the index of the DataFrame that was sampled (result).
  * diamonds.index.isin(result.index) creates a boolean mask where each element is True if the corresponding index is present in the sampled DataFrame (result), and False otherwise.
~diamonds.index.isin(result.index):

  * <P> The tilde (~) is a bitwise NOT operator. It negates the boolean values in the mask. So, ~diamonds.index.isin(result.index) gives a boolean mask where each element is False if the corresponding index is present in the sampled DataFrame (result), and True otherwise. </p>

* diamonds.loc[~diamonds.index.isin(result.index), :]:

  * diamonds.loc[] is used for label-based indexing in Pandas.
  * The boolean mask ~diamonds.index.isin(result.index) is used to select rows from the original DataFrame where the index is not present in the sampled DataFrame.
  * The colon (:) after the comma is used to select all columns.
  * Putting it all together, this expression retrieves rows from the original DataFrame (diamonds) where the index is not present in the sampled DataFrame (result).
  * It effectively gives you the remaining 25% of rows that were not included in the sampled set.
